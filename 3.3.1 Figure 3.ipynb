# 目的：通过签入数量分布，观察高度重复的签入模式

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 导入数据集

```python
df = pd.read_csv("C:/Users/lx/Desktop/Low-MScThesis/BAReport - Dataset/foursquare-nyc-and-tokyo-check-ins/dataset_TSMC2014_TKY.csv")
```

# 统计分析

```python
df1 = df
```

提取每个用户的签入频率

```python
df2 = df1['userId']
```

```python
df2
```

按签入频率对数据集进行排序

```python
df3 = df2.value_counts(ascending=True)
```

```python
df3
```

将数据集放入字典中

```python
df4=dict(df3)
df4
```

将数据集放入列表中

```python
df5 = list(df4.values())
df5
```

删除掉统计范围值外的数据

```python
for i in df5:
    if i > 1500:
        del i
```

# 作出图表

```python
bins=np.arange(50,1450,25)
plt.hist(df5,bins,facecolor="b")
plt.xlabel("Number of Check-Ins Made")
plt.ylabel("Number of Users")
x1 = range(50,1550,50)
y1 = range(0,800,100)
plt.xticks(x1,rotation=90)
plt.yticks(y1)
plt.rcParams['figure.figsize']=(12.0,8.0)
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
plt.show()
```

# 结果分析：上图显示了每个用户的签入数量分布。所有用户在整个期间执行了100多项签入操作，大约80%的人的签入记录不到300。一小部分用户生成了大量签入。对于这组超链接活动用户，我们可以期望观察到高度重复的签入模式
